```python
# Define a function to categorize crimes
def categorize_crime(description):
    if any(keyword in description for keyword in ['BURGLARY', 'THEFT', 'STOLEN', 'SHOPLIFTING', 'VANDALISM', 'ARSON']):
        return 'Property Crime'
    elif any(keyword in description for keyword in ['ASSAULT', 'ROBBERY', 'HOMICIDE', 'BATTERY', 'KIDNAPPING']):
        return 'Violent Crime'
    elif any(keyword in description for keyword in ['RAPE', 'SEXUAL', 'INDECENT', 'CHILD PORNOGRAPHY']):
        return 'Sex Crime'
    elif any(keyword in description for keyword in ['IDENTITY', 'EMBEZZLEMENT', 'FRAUD', 'FORGERY', 'COUNTERFEIT']):
        return 'Fraud & Financial Crime'
    elif 'DRUGS' in description:
        return 'Drug-Related Crime'
    elif any(keyword in description for keyword in ['WEAPON', 'FIREARMS', 'SHOTS FIRED', 'BRANDISH']):
        return 'Weapon-Related Crime'
    elif any(keyword in description for keyword in ['TRESPASSING', 'DISTURBING', 'FALSE POLICE', 'PROWLER']):
        return 'Public Order Crime'
    else:
        return 'Other Crime'

df['crime_category'] = df['crime_code_description'].apply(categorize_crime)

df['crime_category'].value_counts()
```

```python
df = df[~df['crime_category'].isin(['Fraud & Financial Crime', 'Drug-Related Crime'])]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram of target variable
plt.figure(figsize=(10,6))
sns.histplot(df['time_of_occurrence_minutes'], bins=30, kde=True)
plt.title('Distribution of Time of Occurrence (Minutes)')
plt.xlabel('Time of Occurrence (Minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histograms for numerical features
numerical_columns = ['victim_age', 'latitude', 'longitude', 'year', 'month', 'day_of_week', 'day_of_month']
df[numerical_columns].hist(figsize=(12, 10), bins=30)
plt.suptitle('Distribution of Numerical Features')
plt.show()
```

```python
# Plot bar plots for categorical features
categorical_columns = ['victim_sex_F', 'victim_sex_M', 'victim_sex_Unknown', 'crime_category_Other Crime', 'crime_category_Property Crime', 
                       'crime_category_Public Order Crime', 'crime_category_Sex Crime', 'crime_category_Violent Crime', 'crime_category_Weapon-Related Crime']
plt.figure(figsize=(12, 8))

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col)
    plt.title(f'Count of {col}')
    
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplots for identifying outliers
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot of {col}')
    
plt.tight_layout()
plt.show()
```

```python
# Scatter plots for numerical features vs. target variable
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.scatterplot(data=df, x=col, y='time_of_occurrence_minutes')
    plt.title(f'{col} vs Time of Occurrence')
    
plt.tight_layout()
plt.show()
```
